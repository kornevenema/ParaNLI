```python
from bert_score import score
import json
from google.colab import drive

drive.mount('/content/drive')
```

```python
def source_sentences_to_list(path):
  out = []
  with open(path) as f:
    for row in f:
      data = json.loads(row)
      out.append(data["premise"])
      out.append(data["hypothesis"])
  return out
```

```python
def scores_to_file(path, P, R, F1):
  with open(path, "w") as f:
    for i in range(0, len(P)):
      json.dump({"precision": P[i].item(), "recall": R[i].item(), "F1": F1[i].item()}, f)
      f.write('\n')
```

```python
def get_scores(refs_path, cands_path, out_path):
  # original
  refs = source_sentences_to_list(refs_path)
  # a list of model outputs to be evaluataed
  cands = source_sentences_to_list(cands_path)

  P, R, F1 = score(cands, refs, lang='en', verbose=True)
  scores_to_file(out_path, P, R, F1)
```

```python
get_scores("/content/drive/MyDrive/thesis/multinli_dev_matched.jsonl", 
           "/content/drive/MyDrive/thesis/multinli_dev_matched_gpt.jsonl", 
           "/content/drive/MyDrive/thesis/multinli_dev_matched_gpt_bertscore.jsonl")

get_scores("/content/drive/MyDrive/thesis/multinli_dev_mismatched.jsonl", 
           "/content/drive/MyDrive/thesis/multinli_dev_mismatched_gpt.jsonl", 
           "/content/drive/MyDrive/thesis/multinli_dev_mismatched_gpt_bertscore.jsonl")
```
